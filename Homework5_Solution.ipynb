### <font color = 'blue'>F.</font> 

##### <font color = 'blue'>Suppose we classify death rates at or below 10 as 'low', 10+ to 20 as 'medium' and above 20 as 'high'. Identify the countries that fall into each category (no need to print the entire list) and calculate the average and standard deviation of the death rates in each category as well as the number (count) of countries in each category. Display each statistic for each group with a user-friendly message (see the output example below that uses markdown for the header row and code to print results, and round values where needed). Create multiple markdown and code cells to organize your work. The example below does not reflect accurate values from the data and shows only one of the three cases that you must handle.</font>

**Low Death Rates (<=10)**
```
Mean death rate = 5.23
Std dev of death rate = 2.54
Count of countries = 10
```

<HR/>

````python
# Initialized an empty list to classify each row as Low, Medium, or High
DeathRate_Group = []
for i in impdata.index:
    if impdata['DeathRate'][i] <= 10:
        DeathRate_Group.append("Low")
    elif impdata['DeathRate'][i] <= 20:
        DeathRate_Group.append("Medium")
    else:
        DeathRate_Group.append("High")

# Add the list to the dataframe and segregate the Low, Midium and High Deathrate countries into separate dataframes
modified_df = impdata
modified_df['DeathRate_Bucket']= DeathRate_Group
low_df = modified_df[modified_df['DeathRate_Bucket']== "Low"]
medium_df = modified_df[modified_df['DeathRate_Bucket']== "Medium"]
high_df = modified_df[modified_df['DeathRate_Bucket']== "High"]
````

**Low Death Rates (<=10)**

````python
# Printing the Mean, Standard Deviation and Number of Countries in the segregated dataframes
print(f"Mean Death Rate = {round(low_df['DeathRate'].mean(),2)}")
print(f"Standard Deviation of Death Rate = {round(low_df['DeathRate'].std(),2)}")
print(f"Number of Countries = {len(low_df['DeathRate'])}")
````

**Medium Death Rates (10+ to 20)**

````python
print(f"Mean Death Rate = {round(medium_df['DeathRate'].mean(),2)}")
print(f"Standard Deviation of Death Rate = {round(medium_df['DeathRate'].std(),2)}")
print(f"Number of Countries = {len(medium_df['DeathRate'])}")
````

**High Death Rates (>20)**

````python
print(f"Mean Death Rate = {round(high_df['DeathRate'].mean(),2)}")
print(f"Standard Deviation of Death Rate = {round(high_df['DeathRate'].std(),2)}")
print(f"Number of Countries = {len(high_df['DeathRate'])}")
````

**Average life expectancy**

````python
# Calculating the averages of Male and Female Expectancy
MaleExpectancy = round(impdata['MaleLifeExpectancy'].mean(),2)
FemaleExpectancy = round(impdata['FemaleLifeExpectancy'].mean(),2)

# Printing the values
print(f"Avg. male life expectancy = {MaleExpectancy} years")
print(f"Avg. female life expectancy = {FemaleExpectancy} years")

# Using conditional statements to automatically print the realtionship
comparison = ""
if MaleExpectancy> FemaleExpectancy:
    comparison = " greater than"
elif MaleExpectancy == FemaleExpectancy:
    comparison = " equal"
else:
    comparison= " less than"
    
print("Average life expectancy of males is" + comparison + " females.")
````
