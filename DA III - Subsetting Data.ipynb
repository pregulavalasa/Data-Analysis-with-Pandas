### Boolean subsetting

Boolean subsetting can be used to subset rows based on conditions. This type of subsetting can be done for both Series and Data Frames.

```python
# recall how conditions work on variables/ values

gni = 9000
gni > 2000
```

```python
# see how conditions work on columns/ Series
gniData=povData['GNI']
gniData > 2000
```

```python
# filter a series (GNI > 10,000)
gniData[gniData>10000]
```

```python
# Filter a data frame 

# Retrieve all rows where Region is 3
povData[povData['Region']==3]
```

```python
# Retrieve all rows where Region > 3
povData[povData['Region']>3]
```

```python
# Retrieve all rows where Region is 3 or 4
povData[(povData['Region']==3) | (povData['Region']==4)]
```

```python
# Retrieve all rows where Region is between 3 and 5
povData[(povData['Region']>=3) & (povData['Region']<=5)]
```

```python
# Retrieve all rows where Region is not between 3 and 5
povData[~((povData['Region']>=3) & (povData['Region']<=5))]
```

**Note:** Don't forget the parentheses when creating compound conditions. If you do, you will encounter an error.

```python
# List all countries in Region 1 using boolean subsetting followed by column subsetting
povData[povData['Region']==1]['Country']
```

```python
# List all countries in Region 1 using boolean subsetting followed by index property
povData[povData['Region']==1]['Country'].index
```
