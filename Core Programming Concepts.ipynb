## What is a computer program?


A program is a sequence of instructions to perform a specific task. A computer program consists of the sequence of instructions provided in a way that a computer can understand and execute the instructions to accomplish a task.


<BR/>

## What is an algorithm?


An algorithm is a sequence of steps or calculations to be followed to solve a problem or accomplish a task. 

- A computer program implements one or more algorithms to perform a specific task.
- An algorithm or program can be thought of as a recipe. For example, consider the recipe to bake a cake.

    - Heat the oven to 180 degrees celsius.
    
    - Lightly grease a round cake tin with butter.
    
    - Put all ingredients in a mixing bowl and beat until just combined.
    
    - Pour batter into tin and bake for 40 - 50 minutes until a toothpick inserted in the cake comes out clean.

## How does a computer "execute" or "run" a program?


- The computer only understands binary instructions. 

- You can provide the computer binary instructions or instructions in what are known as low-level programming languages like machine code or assembly language. 
    - The instructions in these languages tell the computer which machine operations to perform.   
    - They are quicker to execute but harder to write.

<br/>

- Most people use what are known as high-level programming languages. These are user-oriented in that they are easier for the users to write.
    - The instructions aren't quite in a natural language but the structure is easier for users to understand and code in. 
    - They need to be interpreted or compiled to convert the code written in a high-level language to machine code that the computer can execute. A compiler converts all of the code to machine code before the program executes or runs. An interpreter performs the conversion one line of code at a time while also executing or running the program. 
    - Python is one of the easiest high-level programming languages. It is interpreted.

## What is challenging about programming (Syntax vs Algorithm)?


### Syntax (the vocabulary and grammar of a programming language)
- Learning the syntax is easy! 
- Even if it appears odd at first, this is the easy part because you can easily find the right syntax in books and on the internet (google when you need to). - This is what can be easily *taught* in class. 
- You will get it with a little bit of practice. 
- It does come with some frustration but remember there are many resources to problem-solve this part. 

### Algorithm (the sequence of steps to solve a problem)
- The challenging part is coming up with an algorithm to accomplish a task (problem-solving). 
- This requires computational or algorithmic thinking. This is the part that is harder to *teach* and mostly has to be *learned*. 
- We will cover some concepts that will allow you to start thinking algorithmically but the ONLY way to get better (and quicker) at this is to practice A LOT. Find coding problems and solve them. Most programmers get better at this by practicing a wide-variety of problems and looking at a wide-variety of solutions. 
- There are unfortunately no shortcuts for this. 
- The good news is that problem-solving is a skill that is useful in every aspect of life, not just coding. So you will benefit from it long after this course.
